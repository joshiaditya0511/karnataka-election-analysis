### Adding correct voter turnout for 2023 elections

```python
temp = {
    'const_num':[],
    'constituency':[],
    'turnout':[]
}

ind = []
```

```python
for idx, row in df.iterrows():
    if row['STATE/UT NAME'] == 'TURN OUT':
        ind.append(idx)
```

```python
for idx in ind:
    temp['const_num'].append(
        df.iloc[idx-1]['AC NO.']
    )
    temp['constituency'].append(
        df.iloc[idx-1]['AC NAME']
    )
    temp['turnout'].append(
        df.iloc[idx]['TOTAL']
    )
```

```python
ndf = pd.DataFrame(temp)
```

```python
ndf['const_num'] = ndf.const_num.astype(int)
ndf.turnout = ndf.turnout.astype(int)
```

```python
ndf
```

```python
tdf = pd.read_csv(r"..\..\Final Data\const23.csv", index_col=0)
```

```python
df2 = tdf.merge(ndf, on='const_num')
```

```python
df2 = df2.drop(['constituency_y'], axis='columns').rename({'constituency_x':'constituency'}, axis='columns')
```

```python
df2.to_csv(r"..\..\Final Data\const23.csv")
```

### Extracting all voter data from ECI

```python
import pandas as pd
df = pd.read_excel(r'ECI Results\10-Detailed Results.xlsx', )
```

```python
df.head(5)
```

```python
turnout_idx = df.loc[df['STATE/UT NAME'].str.contains('TURN OUT')].index
```

```python
df.drop(index = turnout_idx, inplace = True)
df.head(5)
```

```python
df = df.drop(
    columns=['STATE/UT NAME', 'GENERAL', 'POSTAL', '% VOTES POLLED', 'SYMBOL']
).rename(
    columns={
        'AC NAME':'constituency',
        'AC NO.':'const_num',
        'CANDIDATE NAME':'name',
        'SEX':'gender',
        'AGE':'age',
        'PARTY':'party',
        'CATEGORY':'constituency_category',
        'TOTAL':'votes',
        'TOTAL ELECTORS':'total_electors',
    }
).astype(
    {
        'constituency':str,
        'const_num':pd.Int64Dtype(),
        'name':str,
        'gender':'category',
        'age':pd.Int64Dtype(),
        'party':'category',
        'constituency_category':'category',
        'votes':pd.Int64Dtype(),
        'total_electors':pd.Int64Dtype(),
    }
)
```

```python
import re

def rmv_dspace(text:str)->str:
    while '  ' in text:
        text = re.sub("  "," ",text)
    return text.strip()

df.name = df.name.str.strip().apply(lambda x: re.sub(r"~|`|!|@|#|\$|%|\^|\&|\*|_|=|\+|\?|\:|\"|\||/|[0-9]",' ',x)).apply(lambda x: re.sub(r"\&",' and ',x)).apply(rmv_dspace)
df.loc[df.name=='Nota','name'] = df.loc[df.name=='Nota', 'name'].str.upper()
```

```python
total_const_votes = df.groupby('const_num')['votes'].sum().reset_index().rename(columns={'votes':'total_const_votes'})
```

```python
df = df.merge(total_const_votes, on='const_num')
```

```python
df.to_csv('../../Final Data/candid_eci23.csv', index=True)
```
